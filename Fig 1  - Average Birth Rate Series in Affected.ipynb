```python
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns 
import scipy.stats as sp

plt.rcParams["font.family"] = "Times New Roman"
plt.figure(figsize=(20,10))
sns.set_style("whitegrid")
#sns.set_palette(sns.color_palette("Greys", 4))
colors = ["#000000", "#696969"]
# Set your custom color palette
sns.set_palette(sns.color_palette(colors))
```

```python
data = mezclados.copy()
```

```python
# Set your custom color palette
sns.set_style("whitegrid")
colors = ["#000000", "#696969"]
sns.set_palette(sns.color_palette(colors))

data['comuna_utf8'] = data["comuna_utf8"].astype('category')
data['afect'] = data['afect'].astype('category')
data['serie'] = data['serie'].astype('category')

data = data[(data.agno>2003) & (data.agno<2016)].copy()
plt.figure(dpi=1200)
#Gráfico general
plt.subplots(figsize=(20,10))
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 20
plt.plot(data.groupby('agno')['natalidad'].mean(), label="Average series")
plt.plot(data[(data.afect==1)].groupby('agno')['natalidad'].mean(), label="Affected zone", linestyle='dashed', lw=1)
plt.plot(data[(data.afect==0)].groupby('agno')['natalidad'].mean(), label="Not affected zone" , linestyle='dotted', lw=1)
plt.title("Average Birth Rate Series in Affected Versus Unaffect Zones (2004-2015)", fontsize=20, fontweight='bold')
plt.suptitle("2010 Chile earthquake", fontsize=20)
plt.legend(loc='lower left')
plt.xlabel("Years")
plt.ylabel("Birth rate")

plt.savefig('img1.png', transparent = True)
plt.savefig('img1.pdf')

plt.show()
```
